# **MatPlotLib**

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 50, 15, 40]
plt.plot(x, y)
plt.show()
```

```python
months = [1,2,3,4]
sales = [1880, 1450, 504, 2504]

plt.plot(months, sales, color='red', linestyle='--', linewidth=2, label='2025 sales report', marker='o')
plt.title('Sales Report')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
