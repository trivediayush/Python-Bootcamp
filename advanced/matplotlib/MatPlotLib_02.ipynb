```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
```

# **Basic Line Plot**

```python
print("=== Line Plot ===")

plt.figure()
plt.plot(x, y, label='sin(x)', color='blue', linestyle='-', marker='')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', marker='')
plt.title("Line Plot: sin(x) and cos(x)")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

# **Scatter Plot**

```python
print("=== Scatter Plot ===")

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 100 * np.random.rand(50)

plt.figure()
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar()
plt.grid(True)
plt.show()
```

# **Bar Plot**

```python
print("=== Bar Plot ===")

categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.figure()
plt.bar(categories, values, color='skyblue')
plt.title("Bar Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

# **Horizontal Bar Plot**

```python
print("=== Horizontal Bar Plot ===")

plt.figure()
plt.barh(categories, values, color='orange')
plt.title("Horizontal Bar Plot")
plt.xlabel("Values")
plt.ylabel("Category")
plt.show()
```

# **Histogram**

```python
print("=== Histogram ===")

data = np.random.randn(1000)

plt.figure()
plt.hist(data, bins=30, color='green', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

# **Pie Chart**

```python
print("=== Pie Chart ===")

sizes = [25, 35, 20, 20]
labels = ['Python', 'Java', 'C++', 'JavaScript']
explode = (0.1, 0, 0, 0)

plt.figure()
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Pie Chart of Languages")
plt.axis('equal')
plt.show()
```

# **Subplots**

```python
print("=== Subplots ===")

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='blue')
plt.title("Sine")

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red')
plt.title("Cosine")

plt.tight_layout()
plt.show()
```

# **Customizations and Styling**

```python
print("=== Customization ===")

plt.figure()
plt.plot(x, y1, color='purple', linewidth=2, linestyle='--', marker='o', markersize=4)
plt.title("Customized Line Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True, linestyle='--', linewidth=0.5)
plt.axhline(0, color='black', linewidth=1)
plt.axvline(5, color='gray', linestyle='--')
plt.show()
```

# **Save Plot to File**

```python
print("=== Save Plot ===")

plt.figure()
plt.plot(x, y1)
plt.title("Saved Plot")
plt.savefig("plot_output.png", dpi=300)
plt.close()  # Close to avoid displaying
```
